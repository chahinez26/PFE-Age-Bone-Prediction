```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models, applications, callbacks
import matplotlib.pyplot as plt
import time
from IPython.display import display, Javascript

# Paths
BASE_DIR = "/content/drive/MyDrive"
csv_path = os.path.join(BASE_DIR, "bone_age_project/data/labels.csv")
images_folder = os.path.join(BASE_DIR, "bone_age_project/data/images")
OUTPUT_DIR = os.path.join(BASE_DIR, 'bone_age_project/Models')

# Ensure the output directory exists
os.makedirs(OUTPUT_DIR, exist_ok=True)

# -----------------------------
# Preprocessing Functions
# -----------------------------
def normalize_intensity(image):
    image = tf.cast(image, tf.float32) / 255.0
    mean = tf.reduce_mean(image)
    std = tf.math.reduce_std(image)
    image = (image - mean) / (std + 1e-5)
    return image

def resize_with_pad(image, target_height=224, target_width=224):
    return tf.image.resize_with_pad(image, target_height, target_width)

def data_augmentation(image):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_brightness(image, 0.1)
    image = tf.image.random_contrast(image, 0.8, 1.2)
    angle = tf.random.uniform([], -10, 10) * (np.pi / 180)
    image = tf.image.rot90(image, k=tf.cast(angle * 2 / 3.14159, tf.int32))
    return image

def preprocess_image(image_path, augment=False):
    image_str = tf.io.read_file(image_path)
    image = tf.image.decode_png(image_str, channels=1)
    image = normalize_intensity(image)
    image = resize_with_pad(image)
    if augment:
        image = data_augmentation(image)
    return image

def create_dataset(dataframe, images_folder, batch_size=32, augment=False, shuffle=True):
    image_paths = [os.path.join(images_folder, str(id) + ".png") for id in dataframe['id']]
    ages = dataframe['boneage'].values / 228.0
    genders = dataframe['male'].values.astype(np.float32)

    dataset = tf.data.Dataset.from_tensor_slices((image_paths, ages, genders))
    dataset = dataset.map(
        lambda path, age, gender: (
            {'image_input': preprocess_image(path, augment), 'gender_input': tf.expand_dims(gender, axis=-1)},
            age
        ),
        num_parallel_calls=tf.data.AUTOTUNE
    )
    if shuffle:
        dataset = dataset.shuffle(1000)
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

# -----------------------------
# Model Building
# -----------------------------
def build_model():
    # Inputs
    image_input = layers.Input(shape=(224, 224, 1), name='image_input')
    gender_input = layers.Input(shape=(1,), name='gender_input')

    # Convert grayscale to 3 channels
    x = layers.Concatenate()([image_input, image_input, image_input])
    base_model = applications.MobileNetV2(input_shape=(224, 224, 3),
                                          include_top=False,
                                          weights='imagenet')
    base_model.trainable = False

    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Concatenate()([x, gender_input])
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    output = layers.Dense(1, activation='linear')(x)

    model = tf.keras.Model(inputs=[image_input, gender_input], outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                  loss='mse',
                  metrics=[tf.keras.metrics.MeanAbsoluteError(name="mae")])
    return model

# -----------------------------
# Training Script
# -----------------------------
def run_training(csv_path, images_folder, batch_size=32, epochs=10):
    df = pd.read_csv(csv_path)
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

    train_dataset = create_dataset(train_df, images_folder, batch_size, augment=True)
    test_dataset = create_dataset(test_df, images_folder, batch_size, augment=False, shuffle=False)

    model = build_model()

    checkpoint_cb = callbacks.ModelCheckpoint("mobilenet_bestV2.h5", save_best_only=True, monitor="val_mae")
    earlystop_cb = callbacks.EarlyStopping(patience=5, restore_best_weights=True)

    history = model.fit(
        train_dataset,
        validation_data=test_dataset,
        epochs=30,
        callbacks=[checkpoint_cb, earlystop_cb]
    )

    # Save the model after training
    model.save(os.path.join(OUTPUT_DIR, 'final_fine_tuned_model_RSNA.h5'))
    print(f"Model saved to {os.path.join(OUTPUT_DIR, 'final_fine_tuned_model_RSNA.h5')}")

    return model, history, test_dataset

# -----------------------------
# Plotting & Evaluation
# -----------------------------
def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Perte (MSE)')
    plt.xlabel('Époques')
    plt.ylabel('Perte')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['mae'], label='Train MAE')
    plt.plot(history.history['val_mae'], label='Val MAE')
    plt.title('Erreur absolue moyenne (MAE)')
    plt.xlabel('Époques')
    plt.ylabel('MAE')
    plt.legend()

    plt.tight_layout()
    plt.show()

    best_epoch = np.argmin(history.history['val_mae'])
    print(f"Meilleure époque: {best_epoch + 1}")
    print(f"Train MAE: {history.history['mae'][best_epoch] * 228:.2f} mois")
    print(f"Val MAE: {history.history['val_mae'][best_epoch] * 228:.2f} mois")

def evaluate_model(model, test_dataset):
    test_loss, test_mae = model.evaluate(test_dataset)
    print(f"\nTest Loss (MSE): {test_loss:.4f}")
    print(f"Test MAE: {test_mae:.4f} → {test_mae * 228:.2f} mois")
    return test_loss, test_mae

def visualize_predictions(model, test_dataset, num_samples=10):
    for inputs, true_ages in test_dataset.take(1):
        pred_ages = model.predict(inputs)
        true_ages_months = true_ages.numpy() * 228
        pred_ages_months = pred_ages.flatten() * 228
        errors = np.abs(true_ages_months - pred_ages_months)

        plt.figure(figsize=(15, 10))
        for i in range(min(num_samples, len(true_ages))):
            plt.subplot(2, 5, i+1)
            plt.imshow(inputs['image_input'][i].numpy().squeeze(), cmap='gray')
            gender = "M" if inputs['gender_input'][i].numpy()[0] == 1 else "F"
            plt.title(f"Vrai: {true_ages_months[i]:.1f} ({gender})\nPrédit: {pred_ages_months[i]:.1f}\nErreur: {errors[i]:.1f}")
            plt.axis('off')
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(8, 6))
        plt.scatter(true_ages_months, pred_ages_months, alpha=0.5)
        plt.plot([0, 228], [0, 228], 'r--')
        plt.xlabel('Âge réel (mois)')
        plt.ylabel('Âge prédit (mois)')
        plt.title('Comparaison des âges')
        plt.grid(True)
        plt.show()

# -----------------------------
# Run everything
# -----------------------------
model, history, test_dataset = run_training(csv_path, images_folder, batch_size=32, epochs=20)
# --- Sauvegarde finale ---
# Cette ligne est désormais incluse dans la fonction `run_training`
plot_training_history(history)
evaluate_model(model, test_dataset)
visualize_predictions(model, test_dataset, num_samples=10)

# --- Code pour maintenir la session active ---


def keep_alive():
    while True:
        display(Javascript('window.dispatchEvent(new Event("resize"))'))  # Force un événement de redimensionnement
        time.sleep(60 * 5)  # Attendre 5 minutes avant de déclencher un autre événement

# Lancer la fonction pour maintenir la session active
keep_alive()
```
